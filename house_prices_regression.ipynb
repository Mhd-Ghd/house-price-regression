```python
import pandas as pd 
import numpy as np
import math
```

#**Importing the dataset**

```python
data = pd.read_csv("train.csv")
x_df = data.loc[:,["MSSubClass","LotArea","OverallQual","OverallCond","YearBuilt","1stFlrSF","2ndFlrSF","GrLivArea","GarageArea","YrSold"]]
x = x_df.values 
y = data.iloc[:,-1].values

```

#**features scalling**

```python
max = x_df.max().values
x = x/max
x_train = np.c_[np.ones(len(x)),x]  ##concatenate a vector of ones to the fetures matrix
y_train = y
x_train
```

#**training the model**

```python
theta = np.random.rand(len(x_train[0]),1)
lr = 0.1
loss = []
print(f"X_trian is : {len(x_train)}x{len(x_train[0])} and theta is {len(theta)}x{len(theta[0])}")
print(theta)
```

```python
prev_loss = -math.inf 
current_loss = math.inf
while (abs(current_loss - prev_loss) > (1**10^-3)):
  y_pred = np.matmul(x_train,theta)
  grad = np.matmul(np.matrix.transpose(x_train) , np.subtract(y_pred,y_train))/len(x_train)
  theta = np.subtract(theta,lr*grad)
  loss.append(np.sum(np.square(np.subtract(y_pred,y_train)))/len(x_train))


print(loss)
```
